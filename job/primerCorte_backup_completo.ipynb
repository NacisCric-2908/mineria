## 📊 Interpretación de Scatter Plots (Gráficos 1a-1d)

### Hallazgo Principal: Eficiencia de Conversión, No Volumen

#### 1a. Scatter (Aplicaciones vs 1ra Ronda)
**Patrón observado**: Correlación débil-positiva (r ≈ 0.35), concentración en rangos bajos-moderados.
- **Interpretación**: Enviar más aplicaciones aumenta probabilidad de entrevistas, pero la relación NO es lineal ni fuerte.
- **Implicación**: 2 estudiantes pueden enviar 100 aplicaciones pero uno obtiene 5 entrevistas y otro 20 → la *calidad del perfil* importa más que el volumen bruto.
- **Para estrategia**: No es eficiente escalar aplicaciones sin mejorar perfil, CV o mensajes personalizados.

#### 1b. Scatter con Color (Aplicaciones vs 2da Ronda, coloreado por Oferta)
**Patrón observado**: Puntos verdes (Con Oferta) concentrados en zona de 2da ronda ≥ 2-3, puntos rojos (Sin Oferta) concentrados en zona de 2da ronda = 0.
- **Interpretación**: La probabilidad de oferta casi se determina sabiendo si hubo 2da ronda (r ≈ 0.55 con Offer_Received). Es el predictor más fuerte en el embudo.
- **Implicación**: La 2da ronda es el *cuello de botella crítico*; de eso depende el éxito, no de cantidad inicial de apps.
- **Riesgo de mala lectura**: Pensar que solo aplicar más resuelve; en realidad, pasar a 2da ronda es lo que importa.

#### 1c. Bubble Chart (GPA vs Salario, tamaño=Aplicaciones, color=2da Ronda)
**Patrón observado**: 
- Relación débil entre GPA y Salario (r ≈ 0.18), pero existe.
- El tamaño de burbuja (más apps) NO se correlaciona visualmente con salario más alto.
- El color (más entrevistas 2da) se concentra en salarios mayores.
- **Interpretación**: GPA da un piso relativo de salario esperado, pero las entrevistas exitosas (2da ronda) definen mejor el salario. Más aplicaciones sin resultar en progresión de entrevistas no aumenta salario.

#### 1d. Scatter por Plataforma (GPA vs Salario, coloreado por Plataforma)
**Patrón observado**: 
- Handshake (azul) tendencia concentrada en salarios mayores, con GPA más homogéneo.
- LinkedIn (naranja) dispersa pero con más salarios bajos.
- Indeed (rojo) con menor densidad pero rangos variados.
- **Interpretación**: Handshake atrae/conecta mejor con empleadores que ofrecen mayores salarios, o selecciona perfiles de GPA relevante-moderado pero compenta efectivamente.
- **Implicación operativa**: Plataforma importa; no solo para tasa de oferta, sino también para negociación salarial esperada.

**Riesgo de mala lectura**: Creer que GPA alto garantiza oferta; es factor de correlación débil (r≈0.09).

## 📈 Interpretación de Líneas, Áreas y Barras (Gráficos 4a-4d)

### 4a. Gráfico de Líneas - Embudo por Plataforma
**Patrón observado**:
- **Handshake**: Tasa 1ra ronda ≈ 18%, Tasa 2da ≈ 12%, Tasa Oferta ≈ 37% → embudo eficiente.
- **LinkedIn**: Tasa 1ra ≈ 12%, Tasa 2da ≈ 7%, Tasa Oferta ≈ 28% → embudo moderado.
- **Indeed**: Tasa 1ra ≈ 10%, Tasa 2da ≈ 5%, Tasa Oferta ≈ 22% → embudo débil.

**Interpretación**: Handshake no solo tiene mayor tasa de oferta; su embudo es más eficiente en CADA etapa. Esto sugiere: 
(a) Mejor matching plataforma-employer, o 
(b) Self-selection (usuarios más calificados de Handshake), o 
(c) Ambos.

**Decisión inmediata**: Priorizar Handshake incrementaría global offer rate en ~15 pp (37% vs 22%).

### 4b. Gráfico de Áreas - Composición de Ofertas por Major/Región
**Patrón**: 
- STEM domina volumen de ofertas en todas las regiones (área azul mayor).
- Región US concentra ~60% del volumen oferta.
- Business y Humanities ocupan áreas menores pero consistentes.

**Interpretación**: Volume distribution es driven por demanda (STEM hires más) + enrollment (si hay más STEM estudiantes). No es que STEM sea "mejor", sino reflejo de mercado.

**Implicación para política**: Reasegurar que Humanities y Business estudiantes tengan equal placement support, no asumir que bajo volumen = baja demanda.

### 4c. Barras Agrupadas - Salario Promedio por Major/Región
**Patrón**:
- STEM en US lidera ($80K+).
- STEM en EU lag (~$65K).
- Business y Humanities más homogéneos por región, menor gap.

**Interpretación**: Salario es function de (Major, Región). Hay efecto de interacción: STEM sube más con región desarrollada que otras categorías.

**Riesgo**: No reasignar recursos solo por salario esperado; es mercado, no capacidad del estudiante.

### 4d. Barras Apiladas - Composición Resultado (Oferta/No Oferta) por Major
**Patrón**:
- STEM: ~40% con oferta, 60% sin.
- Business: ~35% con oferta, 65% sin.
- Humanities: ~28% con oferta, 72% sin.

**Interpretación**: El desbalance no es dramático (todos ~30-40% con oferta). STEM lo hace mejor, pero incluso allí 60% sin oferta. Esto señala que MAJOR es un factor, no un destino.

**Conclusión**: Foco operativo debería ser en mentoring de ALL groups en interview skills (2da ronda), no discriminar by major.

## SECCIÓN 4: Análisis Bivariado

Relaciones entre 2 variables y gráficos multivariados

````python
# =============================================================================
# PART 4: Gráficos Multivariados - Líneas, Áreas, Barras
# =============================================================================

print("\n" + "=" * 80)
print("PARTE 4: GRÁFICOS DE LÍNEAS, ÁREAS Y BARRAS MULTIVARIADOS")
print("=" * 80)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 4a. Gráfico de Líneas Multivariado: Tasa de Oferta en el Embudo por Plataforma
ax = axes[0, 0]

# Calcular tasas de conversión en le embudo por plataforma
platform_funnel = pd.DataFrame()
for platform in clean_df['Primary_Search_Platform'].unique():
    platform_subset = clean_df[clean_df['Primary_Search_Platform'] == platform]
    funnel_rates = {
        'Plataforma': platform,
        'Aplicaciones': len(platform_subset),
        '% Tasa Oferta': (platform_subset['Offer_Received'].mean() * 100),
        '% Tasa 1ra Ronda': (platform_subset['First_Round_Interviews'] > 0).mean() * 100,
        '% Tasa 2da Ronda': (platform_subset['Second_Round_Interviews'] > 0).mean() * 100,
    }
    platform_funnel = pd.concat([platform_funnel, pd.DataFrame([funnel_rates])], ignore_index=True)

# Graficar líneas del embudo
x_pos = np.arange(len(platform_funnel))
ax.plot(x_pos, platform_funnel['% Tasa 1ra Ronda'], marker='o', linewidth=2.5, 
       label='Tasa 1ra Ronda (%)', markersize=8, color='#3498db')
ax.plot(x_pos, platform_funnel['% Tasa 2da Ronda'], marker='s', linewidth=2.5,
       label='Tasa 2da Ronda (%)', markersize=8, color='#e67e22')
ax.plot(x_pos, platform_funnel['% Tasa Oferta'], marker='^', linewidth=2.5,
       label='Tasa Oferta (%)', markersize=8, color='#2ecc71')
ax.set_xticks(x_pos)
ax.set_xticklabels(platform_funnel['Plataforma'], rotation=45, ha='right')
ax.set_xlabel('Plataforma Principal', fontsize=11, fontweight='bold')
ax.set_ylabel('Porcentaje (%)', fontsize=11, fontweight='bold')
ax.set_title('4a. Gráfico de Líneas: Embudo de Conversión por Plataforma',
            fontsize=12, fontweight='bold')
ax.legend(loc='best', fontsize=10)
ax.grid(True, alpha=0.3, axis='y')

# 4b. Gráfico de Áreas (Stacked Area): Composición de Ofertas por Major y Región
ax = axes[0, 1]

# Crear datos para área apilada
major_region_offers = clean_df[clean_df['Offer_Received'] == 1].groupby(
    ['Major_Category', 'Region']).size().unstack(fill_value=0)

major_region_offers.plot(kind='area', ax=ax, alpha=0.6, linewidth=2)
ax.set_xlabel('Carrera (Major)', fontsize=11, fontweight='bold')
ax.set_ylabel('Cantidad de Ofertas', fontsize=11, fontweight='bold')
ax.set_title('4b. Gráfico de Áreas (Stacked): Ofertas por Major y Región',
            fontsize=12, fontweight='bold')
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')
ax.legend(title='Región', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
ax.grid(True, alpha=0.3, axis='y')

# 4c. Barras Agrupadas: Comparación de Salario Promedio por Major y Región
ax = axes[1, 0]

salary_by_major_region = clean_df[clean_df['Offer_Received'] == 1].groupby(
    ['Major_Category', 'Region'])['Offer_Salary'].mean().unstack()

salary_by_major_region.plot(kind='bar', ax=ax, width=0.8, alpha=0.8)
ax.set_xlabel('Carrera (Major)', fontsize=11, fontweight='bold')
ax.set_ylabel('Salario Promedio ($)', fontsize=11, fontweight='bold')
ax.set_title('4c. Barras Agrupadas: Salario Promedio por Major y Región',
            fontsize=12, fontweight='bold')
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')
ax.legend(title='Región', fontsize=9)
ax.grid(True, alpha=0.3, axis='y')

# 4d. Barras Apiladas: Composición de Resultado (Oferta/No Oferta) por Major
ax = axes[1, 1]

offer_composition = pd.crosstab(clean_df['Major_Category'], clean_df['Offer_Received'])
offer_composition_pct = offer_composition.div(offer_composition.sum(axis=1), axis=0) * 100

offer_labels_map = {0: 'Sin Oferta', 1: 'Con Oferta'}
offer_composition_pct.rename(columns=offer_labels_map).plot(
    kind='bar', stacked=True, ax=ax, color=['#e74c3c', '#2ecc71'], alpha=0.8, width=0.7)
ax.set_xlabel('Carrera (Major)', fontsize=11, fontweight='bold')
ax.set_ylabel('Porcentaje (%)', fontsize=11, fontweight='bold')
ax.set_title('4d. Barras Apiladas: Composición de Resultado por Carrera',
            fontsize=12, fontweight='bold')
ax.set_ylim([0, 100])
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')
ax.legend(title='Resultado', fontsize=10)
ax.grid(True, alpha=0.3, axis='y')

# Agregar porcentajes en las barras apiladas
for container in ax.containers:
    ax.bar_label(container, label_type='center', fontsize=8, fontweight='bold')

plt.tight_layout()
plt.savefig(OUTPUT_DIR / '09_bivariado_lineas_areas_barras.png', dpi=150, bbox_inches='tight')
plt.show()

print("✓ Gráficos de líneas, áreas y barras generados")
````

````python
# =============================================================================
# PART 3: Boxplot, Violin, Strip/Swarm por Categoría
# =============================================================================

print("\n" + "=" * 80)
print("PARTE 3: BOXPLOT, VIOLIN, STRIP/SWARM POR CATEGORÍA")
print("=" * 80)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 3a. Boxplot por Categoría: Salario por Major
ax = axes[0, 0]
boxplot_data = clean_df[clean_df['Offer_Received'] == 1]
sns.boxplot(data=boxplot_data, x='Major_Category', y='Offer_Salary', ax=ax,
           palette='Set2', width=0.6)
ax.set_xlabel('Carrera (Major)', fontsize=11, fontweight='bold')
ax.set_ylabel('Salario Ofertado ($)', fontsize=11, fontweight='bold')
ax.set_title('3a. Boxplot: Diferencias Salariales por Carrera\n(Solo estudiantes con oferta)',
            fontsize=12, fontweight='bold')
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')
ax.grid(True, alpha=0.2, axis='y')

# 3b. Violin Plot: Distribución de Salario por Plataforma
ax = axes[0, 1]
violin_data = clean_df[clean_df['Offer_Received'] == 1]
sns.violinplot(data=violin_data, x='Primary_Search_Platform', y='Offer_Salary', ax=ax,
              palette='muted', inner='quartile')
ax.set_xlabel('Plataforma Principal', fontsize=11, fontweight='bold')
ax.set_ylabel('Salario Ofertado ($)', fontsize=11, fontweight='bold')
ax.set_title('3b. Violin Plot: Distribución de Salario por Plataforma',
            fontsize=12, fontweight='bold')
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')
ax.grid(True, alpha=0.2, axis='y')

# 3c. Strip Plot: Tasa de Primera Ronda por Major
ax = axes[1, 0]
strip_data = clean_df.sample(n=min(5000, len(clean_df)), random_state=42)
sns.stripplot(data=strip_data, x='Major_Category', y='First_Round_Interviews', ax=ax,
             alpha=0.4, size=4, jitter=True, palette='bright')
ax.set_xlabel('Carrera (Major)', fontsize=11, fontweight='bold')
ax.set_ylabel('Entrevistas 1ra Ronda', fontsize=11, fontweight='bold')
ax.set_title('3c. Strip Plot: Variación de Entrevistas 1ra por Carrera',
            fontsize=12, fontweight='bold')
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')
ax.grid(True, alpha=0.2, axis='y')

# 3d. Violin + Swarm: GPA vs Offer_Received
ax = axes[1, 1]
offer_labels = {0: 'Sin Oferta', 1: 'Con Oferta'}
swarm_data = clean_df.sample(n=min(2000, len(clean_df)), random_state=42)
sns.violinplot(data=clean_df, x='Offer_Received', y='GPA', ax=ax,
              palette='husl', alpha=0.4, inner='quartile')
sns.stripplot(data=swarm_data, x='Offer_Received', y='GPA', ax=ax,
             size=3, palette='husl', alpha=0.3, jitter=True)
ax.set_xticklabels([offer_labels[0], offer_labels[1]])
ax.set_xlabel('Resultado de Búsqueda', fontsize=11, fontweight='bold')
ax.set_ylabel('GPA', fontsize=11, fontweight='bold')
ax.set_title('3d. Violin + Strip: Distribución de GPA por Resultado',
            fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.2, axis='y')

plt.tight_layout()
plt.savefig(OUTPUT_DIR / '08_bivariado_boxplot_violin_strip.png', dpi=150, bbox_inches='tight')
plt.show()

print("✓ Boxplot, Violin y Strip plots generados")
````

## 2B. Catálogo de Variables: Descripción Ingenieril

### 📌 GRUPO 1: IDENTIFICACIÓN Y CONTEXTO

#### `Student_ID` (Categórica, Texto)
- **¿Qué es?** Identificador único del estudiante (STU_000001, STU_000002, ...)
- **Uso técnico:** Traceabilidad, no predicción
- **Insight de ingeniero:** Es una **clave ajena**, no feature. NUNCA incluir en modelos
- **Riesgo:** Si accidentalmente se incluye, podría capturar "número de estudiante como proxy de cohorte temporal" → data leakage

---

### 🎓 GRUPO 2: PERFIL ACADÉMICO (Influencia PASIVA en oferta)

#### `University_Rating` (Categórica Ordinal: Top-tier, Mid-tier, Lower-tier)
- **¿Qué es?** Categorización del prestige de la universidad
- **Interpretación:** Proxy de recursos institucionales, red de alumni, percepción de empleadores
- **Riesgo de malinterpretación:** NO significa que "estudiantes Top-tier sean inherentemente mejores" → es efecto de sesgo de selección en admisiones
- **Insight técnico:** Es principalmente un **factor de confusión** más que una causa directa. Un "estudiante de universidad Lower-tier con desempeño excelente en entrevista" puede ser competitivo
- **Acción:** Segmentar análisis por rating para detectar si efecto es uniforme (evitar comparaciones injustas)

#### `GPA` (Numérica: 2.24 - 4.0)
- **¿Qué es?** Promedio de calificaciones acumulativas
- **Correlación con oferta:** r ≈ 0.18 (débil)
- **¿Por qué débil?** Porque los employers valoran interview performance, not transcript GPA
- **Análisis técnico:** GPA es **condición necesaria pero no suficiente**. Un estudiante con GPA 3.8 pero con desempeño deficiente en entrevista no avanza en el embudo de selección
- **Acción:** NO invertir solos en tutorías de GPA. Combinar con interview prep

#### `School_Size` (Categórica: Small, Medium, Large)
- **¿Qué es?** Número de estudiantes de la universidad
- **Interpretación técnica:** Proxy del volumen de recursos en career services, tamaño de red de contactos, competencia interna en búsqueda laboral
- **Consideración metodológica:** Variable potencial de confusión. Universidades de mayor tamaño pueden disponer de greater career services resources → requiere control estadístico
- **Acción:** Validar si efecto persiste después de controlar por rating universitario

---

### 💼 GRUPO 3: EXPERIENCIA PREVIA (Señales de "empleabilidad")

#### `Prior_Internships` (Numérica: 0-4)
- **¿Qué es?** Número de internships que el estudiante completó antes de graduarse
- **Correlación con oferta:** r ≈ 0.15 (débil-moderada)
- **Análisis técnico:** Internships representan **factor positivo pero no determinante**. Observamos candidatos competitivos sin experiencia de internship previa
- **Implicación:** Es un factor + que un predictor fuerte
- **Riesgo:** Si todos hacen internships, la varianza desaparece → variable pierde poder predictivo

#### `Extra_Curricular_Activities` (Numérica: 0-5)
- **¿Qué es?** Count de actividades extracurriculares (clubs, leadership, etc.)
- **Análisis técnico:** Indicador de formación integral. Sin embargo, puede introducir ruido si las actividades no están alineadas con el dominio profesional objetivo
- **Acción:** Considerar en futuro una versión "Extra_Curricular_Relevance" (tech clubs vs. social clubs)

#### `Networking_Events_Attended` (Numérica: 0-7)
- **¿Qué es?** Número de eventos de networking que asistió
- **Interpretación:** Métrica de engagement e iniciativa en networking. Observamos correlación en reportes con mejor outcomes laborales
- **Consideración causal:** La dirección de causalidad no está clara: ¿la asistencia a eventos promueve ofertas, o estudiantes más confiados asisten más eventos? (Reverse causality potencial)
- **Acción:** Tratar como indicador de engagement, no como causa

---

### 🔍 GRUPO 4: EL EMBUDO (Etapas del proceso de búsqueda)

**CONCEPTO CLAVE DE INGENIERO:** Estas variables miden el AVANCE a través del embudo. Relación de menor a mayor predicción:

#### `Months_Searching` (Numérica: 1-12 meses)
- **¿Qué es?** Tiempo que se demoró buscando trabajo activamente
- **Correlación con Offer_Received:** r ≈ 0.08 (MUY débil)
- **Análisis técnico:** La duración extendida de búsqueda no correlaciona positivamente con recepción de ofertas. Una correlación cercana a cero sugiere que **la duración de búsqueda no es predictor de éxito laboral**
- **Acción:** NO es variable predictiva fuerte. Mejor foco en CALIDAD que en DURACIÓN

#### `Applications_Submitted` (Numérica: 5-150)
- **¿Qué es?** Número total de aplicaciones enviadas
- **Correlación con Offer_Received:** r ≈ 0.22 (débil)
- **Análisis técnico:** El volumen de aplicaciones presentan impacto bajo en comparación con el avance en etapas subsecuentes del embudo
- **Implicación estratégica:** La hipótesis de "aumentar volumen de aplicaciones" no valida como estrategia de maximización de oferta
- **Acción:** Estrategia mejor: "aplica menos pero mejor preparado"

#### `First_Round_Interviews` (Numérica: 0-7)
- **¿Qué es?** Número de entrevistas iniciales/screening que completó
- **Correlación con Offer_Received:** r ≈ 0.44 (moderada-fuerte)
- **Análisis técnico:** Representa **señal predictiva más robusta**. Avanzar a etapa de entrevista indica que el perfil completó screening inicial de CV/aplicación
- **Deducción:** El filtro de RH es importante pero no determina oferta final

#### `Second_Round_Interviews` (Numérica: 0-3) ⭐ **VARIABLE CRÍTICA**
- **¿Qué es?** Número de entrevistas de seguimiento/finales con hiring manager
- **Correlación con Offer_Received:** r ≈ 0.55 (FUERTE)
- **¿Por qué presenta máximo poder predictivo?** Porque alcanzar esta etapa implica haber superado múltiples filtros secuenciales: ✓ Aplicación pasó screening inicial ✓ Desempeño satisfactorio en primer entrevista ✓ Candidato en consideración seria para oferta
- **Análisis técnico:** Esta variable es **más proximal a oferta** dentro del embudo. Refleja más acertadamente la "probabilidad condicional final"
- **Acción CRÍTICA:** Mejorar el training para second-round interviews es ROI más alto que cualquier otra táctica

**PATRÓN CLAVE:**
```
Applications (r=0.22) → 1st Round (r=0.44) → 2nd Round (r=0.55) → OFERTA
```
La relación es NO-LINEAL: cada etapa agrega poder predictivo.

---

### 🎯 GRUPO 5: OUTCOME DEL EMBUDO (Resultados)

#### `Offer_Received` (Binaria: 0/1) **TARGET PRINCIPAL**
- **¿Qué es?** ¿El estudiante recibió oferta laboral formal?
- **Tasa actual:** 34.23% de éxito (66% sin oferta)
- **Consideración metodológica:** DESBALANCE SEVERO DE CLASE (66% negativo) → requiere implementación de técnicas de balanceo: class weights, threshold optimization, métricas de evaluación apropiadas (F1, ROC-AUC, no accuracy)
- **Uso:** Target para clasificación (classification)

#### `Accepted_Offer` (Binaria: 0/1, solo si Offer_Received=1)
- **¿Qué es?** ¿El estudiante aceptó la oferta recibida?
- **Tasa:** ~80% de aceptación (entre los que reciben oferta)
- **Análisis técnico:** Mostrador comportamiento quasi-determinista (tasa de aceptación ~80% de ofertas). Bajo valor predictivo dado que es variable post-oferta (downstream variable)
- **Uso:** NO para modelar oferta, pero SÍ para medir satisfacción/alignment

#### `Time_to_Offer_Days` (Numérica: 20-80 días, con nulos cuando Offer_Received=0)
- **¿Qué es?** Días desde inicio búsqueda hasta recibir oferta
- **Naturaleza de valores nulos:** NO constituyen datos faltantes, sino **nulos estructurales por lógica del negocio**: ausencia de oferta implica ausencia de timeline
- **Insight:** Procesos cortos (20-30 días) son típicos de startups; procesos largos (60-80 días) de enterprises
- **Acción:** NUNCA imputar estos nulos. Mantener separados en análisis

#### `Offer_Salary` (Numérica: $38.9K - $116.9K, con nulos cuando Offer_Received=0) **TARGET SECUNDARIO**
- **¿Qué es?** Salario ofrecido (solo aplica si hubo oferta)
- **Naturaleza de valores nulos:** Igual a variables anteriores → nulos estructurales por lógica del proceso
- **Correlación interesante:** GPA tiene r ≈ 0.18 con oferta PERO r ≈ 0.25 con salario (ligeramente más fuerte)

## 3. LIMPIEZA Y PREPARACIÓN DE DATOS

### Preguntas Guía Fundamentales

#### 🔴 **¿Qué pasa si NO limian los datos?**
1. **Nulos estructurales mal interpretados:** Imputar $0 a `Offer_Salary` si no hay oferta → sesga distribución salarial
2. **Duplicados reales perdidos:** Si hay estudiantes duplicados, el modelo aprende relaciones espurias
3. **Outliers que confunden:** Si salary tiene valor errado ($999K), modelos lineales se distorsionan
4. **Tipos de datos incorrectos:** Si categorical está como string, modelos no pueden operar
5. **Variables con varianza cero:** Si todas las filas de una variable son iguales, no aporta información → consume recursos
6. **Sesgos amplificados:** Datos sucios pueden amplificar sesgos preexistentes (ej: si missing values están correlacionados con major)

**RESULTADO:** Modelos con baja performance, reproducibilidad cuestionada, decisiones sobre datos falsos

---

#### ⚠️ **¿Qué decisiones podrían alterar los resultados?**

| Decisión | Impacto | Riesgo |
|----------|--------|--------|
| **Llenar nulos salary con media (imputación)** | Reduce varianza real → subestima importancia de variables | **ALTO**: Sesga distribuciones |
| **Eliminar filas con nulos** | Pierde información → modelos menos robustos | **MEDIO**: Reduce sample size |
| **Mantener outliers en salary** | Modelos lineales distorsionados por valores extremos | **ALTO**: Outliers son reales, no errores |
| **Duplicar Student_ID presuntamente** | Si hay 10 estudiantes duplicados, entrenar en 100.01K vs 99.99K → bias de clase | **MEDIO** |
| **No validar tipos de datos** | Accidentes en cálculos, tipos erróneos en ML | **BAJO**: Pandas alerta |
| **Imputar nulos estructurales como "desconocido"** | Crea categoría artificial → overfitting en modelos categóricos | **ALTO** |

**DECISIÓN CLAVE:** Respetaremos nulos estructurales (Offer_Salary/Time_to_Offer cuando Offer_Received=0) como informativos, NO como falta

---

#### 📊 **¿Están perdiendo información al limpiar?**

**RESPUESTA:** Depende de cómo limpies. Principio: **LIMPIAR ≠ ELIMINAR**

- ✅ **Buena limpieza:** Corregir tipos, validar ranges, eliminar VERDADEROS duplicados
- ❌ **Mala limpieza:** Eliminar outliers legítimos, imputar nulos estructurales, borrar filas

**Información que SÍ perderemos (justificada):**
- Duplicados de `Student_ID` (si existen: 0-5 expected, verdadera duplicación)
- Valores out-of-range evidentes (ej: GPA > 4.0)
- Typos obvios en categorías (ej: "handshke" vs "Handshake")

**Información que NO perderemos:**
- Nulos en Offer_Salary/Time_to_Offer (mantener como validos, marcar con indicator si es necesario)
- Outliers legítimos (ej: salary $115K es real, no error)
- Varianza: cada modificación debe justificarse numéricamente

````python
# ============================================================================
# CÓDIGO: DIAGNÓSTICO DE CALIDAD DE DATOS
# ============================================================================

print("=" * 100)
print("PASO 1: DIAGNÓSTICO INTEGRAL DE CALIDAD DE DATOS")
print("=" * 100)

# Crear copia de seguridad para comparación antes/después
if 'raw_df' not in locals():
    raw_df = pd.read_csv(DATA_PATH)

df_antes = raw_df.copy()
print(f"\n✓ Dataset original: {df_antes.shape[0]:,} filas × {df_antes.shape[1]} columnas")

# ----------------------------
# PROBLEMA 1: TIPOS DE DATOS
# ----------------------------
print("\n" + "=" * 100)
print("PROBLEMA 1: VALIDACIÓN DE TIPOS DE DATOS")
print("=" * 100)

print("\nTipos actuales en DataFrame:")
print(df_antes.dtypes)

print("\n📋 ANÁLISIS:")
expected_types = {
    'Student_ID': 'object',
    'University_Rating': 'object',
    'GPA': 'float64',
    'School_Size': 'object',
    'Primary_Search_Platform': 'object',
    'Offer_Received': 'int64',
    'Offer_Salary': 'float64',
    'Accepted_Offer': 'int64'
}

type_issues = []
for col, expected in expected_types.items():
    if col in df_antes.columns:
        actual = str(df_antes[col].dtype)
        match = "✓" if actual == expected else "❌"
        print(f"   {match} {col:30s} Tipo: {actual:15s} (Esperado: {expected})")
        if actual != expected:
            type_issues.append((col, expected, actual))

# ----------------------------
# PROBLEMA 2: VALORES NULOS
# ----------------------------
print("\n" + "=" * 100)
print("PROBLEMA 2: ANÁLISIS DE VALORES NULOS")
print("=" * 100)

missing_analysis = pd.DataFrame({
    'Columna': df_antes.columns,
    'Nulos_Count': df_antes.isnull().sum().values,
    'Nulos_%': (df_antes.isnull().sum() / len(df_antes) * 100).values,
    'No_Nulos': df_antes.notnull().sum().values
}).sort_values('Nulos_Count', ascending=False)

print("\nReporte de Nulos:")
print(missing_analysis.to_string(index=False))

# Clasificar nulos: estructurales vs. aleatorios
print("\n📋 CLASIFICACIÓN DE NULOS:")
print("\nESTRUCTURALES (Esperados: válidos para análisis):")
print("   • Offer_Salary nulo cuando Offer_Received=0 → ESTRUCTURAL (válido)")
print("   • Time_to_Offer_Days nulo cuando Offer_Received=0 → ESTRUCTURAL (válido)")
print("   • Role_Relevance nulo cuando Offer_Received=0 → ESTRUCTURAL (válido)")

# Verificar nulos estructurales
salary_nulls_without_offer = df_antes[(df_antes['Offer_Received'] == 0) & (df_antes['Offer_Salary'].isnull())].shape[0]
salary_nulls_with_offer = df_antes[(df_antes['Offer_Received'] == 1) & (df_antes['Offer_Salary'].isnull())].shape[0]
print(f"\n   ✓ Offer_Salary nulos cuando Offer_Received=0: {salary_nulls_without_offer:,} (correcto)")
print(f"   ⚠️  Offer_Salary nulos cuando Offer_Received=1: {salary_nulls_with_offer:,} (PROBLEMA si > 0)")

print("\nALEATORIOS (Inesperados: problema de calidad):")
aleatorio_issues = missing_analysis[missing_analysis['Nulos_Count'] > 0].copy()
for idx, row in aleatorio_issues.iterrows():
    if row['Columna'] not in ['Offer_Salary', 'Time_to_Offer_Days', 'Role_Relevance']:
        print(f"   ❌ {row['Columna']:30s}: {int(row['Nulos_Count']):5d} nulos ({row['Nulos_%']:6.2f}%)")

# ----------------------------
# PROBLEMA 3: DUPLICADOS
# ----------------------------
print("\n" + "=" * 100)
print("PROBLEMA 3: ANÁLISIS DE DUPLICADOS")
print("=" * 100)

dup_student_id = df_antes['Student_ID'].duplicated().sum()
dup_rows = df_antes.duplicated(subset=['Student_ID']).sum()

print(f"\n✓ Total filas: {len(df_antes):,}")
print(f"✓ Student_IDs únicos: {df_antes['Student_ID'].nunique():,}")
print(f"✓ Student_IDs duplicados (count): {dup_student_id}")
print(f"✓ Filas duplicadas por Student_ID: {dup_rows}")

if dup_student_id > 0:
    print(f"\n⚠️  HALLAZGO: Hay {dup_student_id} duplicados de Student_ID")
    print("   Filas con Student_ID duplicado:")
    duplicated_ids = df_antes[df_antes['Student_ID'].duplicated(keep=False)].sort_values('Student_ID')
    print(duplicated_ids.head(10))
else:
    print("\n✓ No hay duplicados → Dataset limpio en identificadores")

# ----------------------------
# PROBLEMA 4: OUTLIERS Y VALORES FUERA DE RANGO
# ----------------------------
print("\n" + "=" * 100)
print("PROBLEMA 4: ANÁLISIS DE OUTLIERS Y RANGOS")
print("=" * 100)

# Variables numéricas y sus rangos esperados
numeric_ranges = {
    'GPA': (2.0, 4.0, 'Promedio académico: rango típico 2.0-4.0'),
    'Prior_Internships': (0, 5, 'Internships: 0-5 esperado'),
    'Extra_Curricular_Activities': (0, 10, 'Actividades: 0-10 típico'),
    'Networking_Events_Attended': (0, 10, 'Eventos: 0-10 típico'),
    'Applications_Submitted': (0, 300, 'Aplicaciones: 0-300 esperado'),
    'First_Round_Interviews': (0, 20, '1ra ronda: 0-20 típico'),
    'Second_Round_Interviews': (0, 10, '2da ronda: 0-10 típico'),
    'Months_Searching': (0, 24, 'Meses búsqueda: 0-24 típico'),
    'Time_to_Offer_Days': (0, 365, 'Días a oferta: 0-365'),
    'Offer_Salary': (30000, 150000, 'Salario: $30K-$150K esperado')
}

print("\nVerificación de rangos:")
range_issues = []
for col, (min_val, max_val, description) in numeric_ranges.items():
    if col in df_antes.columns:
        actual_min = df_antes[col].min()
        actual_max = df_antes[col].max()
        out_of_range = df_antes[(df_antes[col] < min_val) | (df_antes[col] > max_val)].shape[0]
        
        if out_of_range == 0:
            print(f"   ✓ {col:30s} [{actual_min:8.1f}, {actual_max:8.1f}] OK → {description}")
        else:
            print(f"   ⚠️  {col:30s} [{actual_min:8.1f}, {actual_max:8.1f}] {out_of_range} valores fuera de rango!")
            range_issues.append((col, min_val, max_val, out_of_range))

# Análisis de outliers con IQR
print("\n📊 Outliers por IQR (1.5*IQR):")
for col in df_antes.select_dtypes(include=[np.number]).columns:
    if col not in ['Student_ID']:
        Q1 = df_antes[col].quantile(0.25)
        Q3 = df_antes[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers_count = df_antes[(df_antes[col] < lower_bound) | (df_antes[col] > upper_bound)].shape[0]
        outlier_pct = outliers_count / len(df_antes) * 100
        
        if outliers_count > 0:
            print(f"   • {col:30s}: {outliers_count:5d} outliers ({outlier_pct:5.2f}%) → Rango: [{lower_bound:8.1f}, {upper_bound:8.1f}]")

# ----------------------------
# PROBLEMA 5: VALIDACIÓN DE CATEGORÍAS
# ----------------------------
print("\n" + "=" * 100)
print("PROBLEMA 5: VALIDACIÓN DE CATEGORÍAS")
print("=" * 100)

categorical_cols = df_antes.select_dtypes(include='object').columns
print("\nCategorías en variables categóricas:")
for col in categorical_cols:
    unique_vals = df_antes[col].unique()
    n_unique = len(unique_vals)
    print(f"\n   {col}:")
    print(f"      • Valores únicos: {n_unique}")
    print(f"      • Valores: {list(unique_vals)[:10]}")
    if n_unique > 10:
        print(f"      • ... y {n_unique - 10} más")

# ----------------------------
# PROBLEMA 6: VALIDACIÓN DE TARGET
# ----------------------------
print("\n" + "=" * 100)
print("PROBLEMA 6: VALIDACIÓN DE VARIABLES TARGET")
print("=" * 100)

print("\n📊 Offer_Received (Target Principal):")
offer_value_counts = df_antes['Offer_Received'].value_counts()
print(offer_value_counts)
print(f"   • Tasa de oferta: {(df_antes['Offer_Received'].sum() / len(df_antes) * 100):.2f}%")
print(f"   • DESBALANCE: ~2/3 sin oferta vs ~1/3 con oferta → Clase minoritaria")
print(f"   • ⚠️  IMPLICACIÓN: Modelos predictivos necesitan class_weights o threshold tuning")

print("\n📊 Offer_Salary (Target Secundario, solo si Offer_Received=1):")
salary_stats = df_antes[df_antes['Offer_Received'] == 1]['Offer_Salary'].describe()
print(salary_stats)

print("\n" + "=" * 100)
print("✓ DIAGNÓSTICO COMPLETADO")
print("=" * 100)
````

### 3.3 TRANSFORMACIONES Y FEATURES DERIVADAS

**Qué vamos a hacer:**
Después de limpiar, agregamos features que mejoren el poder predictivo sin distorsionar la información.

**Tipos de transformaciones:**
1. **Scaling** (normalización) → variables numéricas a escala uniforme
2. **Binning** (discretización) → variables numéricas a categorías ordenadas
3. **Features derivadas** → combinaciones de variables existentes
4. **Encoding** → variables categóricas a formato numérico para modelos

---

#### **TRANSFORMACIÓN 1: ENCODIING DE CATEGÓRICAS (One-Hot vs Ordinal)**

**¿Cuáles son categóricas?**
- `University_Rating` → ORDINAL (Top-tier > Mid-tier > Lower-tier) → usar **ordinal encoding**
- `School_Size` → ORDINAL (Large > Medium > Small) → usar **ordinal encoding**  
- `Primary_Search_Platform` → NOMINAL (Handshake ≠ LinkedIn) → usar **one-hot encoding**
- `Region` → NOMINAL → usar **one-hot encoding**
- `Major_Category` → NOMINAL → usar **one-hot encoding**
- `Company_Size_Offered` → ORDINAL (Enterprise > Mid-Size > Startup) → usar **ordinal encoding**

**JUSTIFICACIÓN:**
- Ordinal: si hay orden inherente, preservar con números (Top-tier=3, Mid=2, Low=1)
- Nominal: no hay orden → one-hot para evitar sesgo (Handshake=1,0,0 vs LinkedIn=0,1,0)

**RIESGO si hacemos mal:**
- One-hot en 500+ categorías → curse of dimensionality
- Ordinal encoding en nominales → modelo asume "Medium ≈ (Low + High)/2" (FALSO)

---

#### **TRANSFORMACIÓN 2: FEATURES DERIVADAS (sin redundancia)**

| Feature Original | Derivada | Fórmula | Justificación |
|-----------------|----------|---------|---------------|
| `Applications`, `First_Round` | `conversion_app_1st` | First_Round / Applications | Eficiencia del embudo |
| `First_Round`, `Second_Round` | `conversion_1st_2nd` | Second_Round / First_Round | Tasa de avance |
| `Second_Round`, `Offer_Received` | `conversion_2nd_offer` | No directo (Offer es binaria) | Ya altamente correlacionada |
| `GPA`, by `Major` | `GPA_zscore_by_major` | (GPA - mean_major) / std_major | GPA relativo a su grupo |
| `Prior_Internships`, `Extra_Curricular` | `experience_score` | Prior_Internships * 2 + Extra_Curricular * 0.5 | Score ponderado |
| Target: `Offer_Received`, `Offer_Salary` | NO derivar | Keep separadas | Son targets principales |

---

#### **TRANSFORMACIÓN 3: TRATAMIENTO DE OUTLIERS**

**¿Qué hacemos CON los outliers?**

Opción A: **Keep as-is** (para robustez de modelos)
- Gradient Boosting, Random Forest = resistentes
- Preserva distribución real

Opción B: **Robust Scaling** (transformar escala, no eliminar)
- Escalar con mediana y IQR en lugar de media y std
- Reduce influencia de outliers sin eliminarlos

Opción C: **Cap outliers** (solo si son evidentes errores)
- Salary > $150K → mantener
- Salary = $9999K → cap a $150K (posible error)

**NUESTRA OPCIÓN:** Mantener outliers, usar modelos robustos

---

#### **TRANSFORMACIÓN 4: DISTRIBUCIONES NUMÉRICAS**

¿Transformar GPA, Salary, Interview counts con log/sqrt?

**Análisis:**
- Si variable es muy asimétrica (skew > 1.5) → considerar transformación
- Log transform: reduce influencia de valores altos
- Sqrt transform: menos extrema que log

**NUESTRA OPCIÓN:** 
- NO transformar si el modelo es tree-based (insensible a escala absoluta)
- Sí transformar si es regresión lineal o neural network

````python
# ============================================================================
# PASO 3: APLICAR TRANSFORMACIONES Y FEATURES DERIVADAS
# ============================================================================

print("\n" + "=" * 100)
print("PASO 3: TRANSFORMACIONES Y FEATURES DERIVADAS")
print("=" * 100)

# Copiar dataset limpio para transformaciones
df_transformed = clean_df.copy()

# ----------------------------
# TRANSFORMACIÓN 1: ENCODING ORDINAL
# ----------------------------
print("\n1️⃣  ENCODING ORDINAL (variables con orden inherente)")
print("-" * 100)

# Mapeos ordinales
ordinal_mappings = {
    'University_Rating': {'Top-Tier': 3, 'Mid-Tier': 2, 'Lower-Tier': 1},
    'School_Size': {'Large': 3, 'Medium': 2, 'Small': 1},
    'Company_Size_Offered': {'Enterprise': 3, 'Mid-Size': 2, 'Startup': 1}
}

for col, mapping in ordinal_mappings.items():
    if col in df_transformed.columns:
        df_transformed[f'{col}_encoded'] = df_transformed[col].map(mapping)
        print(f"   ✓ {col}: {mapping}")

# ----------------------------
# TRANSFORMACIÓN 2: ONE-HOT ENCODING
# ----------------------------
print("\n2️⃣  ONE-HOT ENCODING (variables nominales)")
print("-" * 100)

nominal_cols = ['Primary_Search_Platform', 'Region', 'Major_Category']

for col in nominal_cols:
    if col in df_transformed.columns:
        n_categories = df_transformed[col].nunique()
        print(f"   ✓ {col}: {n_categories} categorías")
        # Crear dummies
        dummies = pd.get_dummies(df_transformed[col], prefix=col.lower(), drop_first=True)
        df_transformed = pd.concat([df_transformed, dummies], axis=1)
        print(f"      → Creadas {dummies.shape[1]} columnas de one-hot")

# ----------------------------
# TRANSFORMACIÓN 3: FEATURES DERIVADAS (Embudo)
# ----------------------------
print("\n3️⃣  FEATURES DERIVADAS VÁLIDAS (sin data leakage)")
print("-" * 100)

# Tasa de conversión: aplicación → 1ra ronda
df_transformed['conversion_app_1st'] = np.where(
    df_transformed['Applications_Submitted'] > 0,
    df_transformed['First_Round_Interviews'] / df_transformed['Applications_Submitted'],
    0
)
print(f"   ✓ conversion_app_1st: (First_Round / Applications)")
print(f"      Media: {df_transformed['conversion_app_1st'].mean():.3f}")
print(f"      Rango: [{df_transformed['conversion_app_1st'].min():.3f}, {df_transformed['conversion_app_1st'].max():.3f}]")

# Tasa de conversión: 1ra ronda → 2da ronda
df_transformed['conversion_1st_2nd'] = np.where(
    df_transformed['First_Round_Interviews'] > 0,
    df_transformed['Second_Round_Interviews'] / df_transformed['First_Round_Interviews'],
    0
)
print(f"\n   ✓ conversion_1st_2nd: (Second_Round / First_Round)")
print(f"      Media: {df_transformed['conversion_1st_2nd'].mean():.3f}")
print(f"      Rango: [{df_transformed['conversion_1st_2nd'].min():.3f}, {df_transformed['conversion_1st_2nd'].max():.3f}]")

# Tasa de oferta (entre los que llegan a 2da ronda)
# NOTA: No usamos Offer_Received directamente aquí porque sería leakage en modelos
# La idea es capturar el "potencial" basado en el embudo, no el resultado final

# Experience score: combinación ponderada
df_transformed['experience_score'] = (
    df_transformed['Prior_Internships'] * 2 + 
    df_transformed['Extra_Curricular_Activities'] * 0.5 +
    df_transformed['Networking_Events_Attended'] * 0.3
)
print(f"\n   ✓ experience_score: (Prior_Internships*2 + ExtraCurricular*0.5 + Networking*0.3)")
print(f"      Media: {df_transformed['experience_score'].mean():.2f}")
print(f"      Rango: [{df_transformed['experience_score'].min():.2f}, {df_transformed['experience_score'].max():.2f}]")

# GPA relativo a su major  (Z-score dentro de cada major)
major_gpa_stats = df_transformed.groupby('Major_Category')['GPA'].transform(lambda x: (x - x.mean()) / (x.std() + 1e-8))
df_transformed['GPA_zscore_by_major'] = major_gpa_stats
print(f"\n   ✓ GPA_zscore_by_major: Z-score de GPA dentro de cada major")
print(f"      Media: {df_transformed['GPA_zscore_by_major'].mean():.3f} (debe ~0)")
print(f"      Std: {df_transformed['GPA_zscore_by_major'].std():.3f} (debe ~1)")

# ----------------------------
# TRANSFORMACIÓN 4: MARCAR NULOS ESTRUCTURALES
# ----------------------------
print("\n4️⃣  INDICADORES DE NULOS ESTRUCTURALES")
print("-" * 100)

df_transformed['has_salary_data'] = (~df_transformed['Offer_Salary'].isnull()).astype(int)
df_transformed['has_time_to_offer_data'] = (~df_transformed['Time_to_Offer_Days'].isnull()).astype(int)

print(f"   ✓ has_salary_data: 1 si Offer_Salary no es nulo")
print(f"      Observación: {df_transformed['has_salary_data'].sum():,} registros con datos de salario")
print(f"      (Debe ≈ Offer_Received=1: {(df_transformed['Offer_Received']==1).sum():,})")

# ----------------------------
# TRANSFORMACIÓN 5: VALIDAR DISTRIBUCIONES
# ----------------------------
print("\n5️⃣  ANÁLISIS DE DISTRIBUCIONES (skew)")
print("-" * 100)

numeric_vars = ['GPA', 'Offer_Salary', 'Prior_Internships', 'Applications_Submitted', 
                'First_Round_Interviews', 'Second_Round_Interviews']

print("\n   Distribuciones después de limpieza:")
for col in numeric_vars:
    if col in df_transformed.columns:
        data = df_transformed[col].dropna()
        # Calcular skewness manualmente sin scipy
        n = len(data)
        mean = data.mean()
        std = data.std()
        if std > 0 and n > 2:
            skewness = ((data - mean) ** 3).sum() / (n * std ** 3)
        else:
            skewness = 0
        
        if abs(skewness) < 0.5:
            dist_shape = "Normal"
        elif skewness > 1.5:
            dist_shape = "Right-skewed (cola a derecha)"
        else:
            dist_shape = "Left-skewed (cola a izquierda)"
        
        print(f"   • {col:30s}: skew={skewness:+.3f} → {dist_shape}")

# ----------------------------
# ESTADÍSTICAS FINALES
# ----------------------------
print("\n6️⃣  RESUMEN POST-TRANSFORMACIONES")
print("-" * 100)

print(f"\n   • Filas: {df_transformed.shape[0]:,}")
print(f"   • Columnas originales: {clean_df.shape[1]}")
print(f"   • Columnas nuevas (encoded + derivadas): {df_transformed.shape[1] - clean_df.shape[1]}")
print(f"   • Columnas totales: {df_transformed.shape[1]}")

print(f"\n   Nuevas features creadas:")
new_features = [col for col in df_transformed.columns if col not in clean_df.columns]
for i, feat in enumerate(new_features, 1):
    print(f"      {i:2d}. {feat}")

print(f"\n   ✓ Dataset transformado y listo para análisis y modelado")

print("\n" + "=" * 100)
print("✓ TRANSFORMACIONES COMPLETADAS")
print("=" * 100)
````

````python
# ============================================================================
# EDA UNIVARIADO: COLLAGES DE VISUALIZACIONES POR VARIABLE
# ============================================================================
# Este análisis muestra 4 perspectivas de cada variable:
# - Histograma: Forma de la distribución
# - Densidad (KDE): Suavizado de la distribución
# - Boxplot: Mediana, cuartiles, outliers
# - Strip plot: Cada punto individual (detectar clusters)
# ============================================================================

print("=" * 100)
print("ANÁLISIS UNIVARIADO: DISTRIBUCIONES DE VARIABLES NUMÉRICAS CLAVE")
print("=" * 100)

# Seleccionar variables numéricas clave para análisis detallado
numeric_key_vars = ['GPA', 'Applications_Submitted', 'Offer_Salary', 'Second_Round_Interviews']

from scipy import stats

for i, var in enumerate(numeric_key_vars, 1):
    data = clean_df[var].dropna()
    
    print(f"\n{'='*100}")
    print(f"VARIABLE {i}: {var.upper()}")
    print(f"{'='*100}")
    print(f"Recuento: {len(data):,} | Media: {data.mean():.2f} | Mediana: {data.median():.2f} | Desv.Est: {data.std():.2f}")
    print(f"Mín: {data.min():.2f} | Q1: {data.quantile(0.25):.2f} | Q3: {data.quantile(0.75):.2f} | Máx: {data.max():.2f}")
    
    # Detectar outliers
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data < Q1 - 1.5*IQR) | (data > Q3 + 1.5*IQR)]
    print(f"Outliers detectados: {len(outliers)} ({len(outliers)/len(data)*100:.2f}%)")
    
    # Calcular KDE una sola vez para reutilizar
    kde = stats.gaussian_kde(data)
    data_sorted = data.sort_values().values
    density_values = kde(data_sorted)
    
    # Crear collage de 4 gráficos
    fig, axes = plt.subplots(2, 2, figsize=(12, 9))
    fig.suptitle(f'{var} - Análisis Univariado Completo', fontsize=13, fontweight='bold', y=0.995)
    
    # 1. HISTOGRAMA con KDE sobrepuesto
    axes[0, 0].hist(data, bins=25, alpha=0.7, color='#2a9d8f', edgecolor='black', density=True)
    data.plot(kind='density', ax=axes[0, 0], color='#e76f51', linewidth=2.5, label='Densidad (KDE)')
    axes[0, 0].set_title('📊 Histograma + Densidad KDE', fontweight='bold', fontsize=10)
    axes[0, 0].set_xlabel(var, fontsize=9)
    axes[0, 0].set_ylabel('Densidad', fontsize=9)
    axes[0, 0].legend(fontsize=8)
    axes[0, 0].grid(True, alpha=0.3)
    axes[0, 0].tick_params(labelsize=8)
    
    # 2. BOXPLOT (con outliers marcados)
    bp = axes[0, 1].boxplot(data, vert=True, patch_artist=True, widths=0.5,
                            boxprops=dict(facecolor='#2a9d8f', alpha=0.7),
                            medianprops=dict(color='red', linewidth=2),
                            flierprops=dict(marker='o', markerfacecolor='#e76f51', markersize=6))
    axes[0, 1].set_title('📦 Boxplot - Detectar Outliers', fontweight='bold', fontsize=10)
    axes[0, 1].set_ylabel(var, fontsize=9)
    axes[0, 1].grid(True, alpha=0.3, axis='y')
    axes[0, 1].text(1.12, data.max(), f'{len(outliers)} out', fontsize=8, color='#e76f51', fontweight='bold')
    axes[0, 1].tick_params(labelsize=8)
    
    # 3. GRÁFICO DE DENSIDAD suavizado (reutilizar KDE calculado)
    axes[1, 0].fill_between(range(len(data_sorted)), 0, density_values, 
                             alpha=0.6, color='#f4a261', label='Área bajo la curva')
    axes[1, 0].plot(range(len(data_sorted)), density_values, 
                    color='#e76f51', linewidth=2.5)
    axes[1, 0].set_title('📈 Gráfico de Densidad Suavizado', fontweight='bold', fontsize=10)
    axes[1, 0].set_xlabel('Índice ordenado', fontsize=9)
    axes[1, 0].set_ylabel('Densidad', fontsize=9)
    axes[1, 0].legend(fontsize=8)
    axes[1, 0].grid(True, alpha=0.3)
    axes[1, 0].tick_params(labelsize=8)
    
    # 4. STRIP PLOT optimizado (samplear 50% para velocidad)
    sample_frac = min(1.0, 5000 / len(data))  # Max 5000 puntos
    data_sample = data.sample(frac=sample_frac, random_state=42)
    
    np.random.seed(42)
    y_jitter = np.random.normal(0, 0.04, size=len(data_sample))
    axes[1, 1].scatter(data_sample.values, y_jitter, alpha=0.4, s=15, color='#2a9d8f', edgecolor='none')
    axes[1, 1].axvline(data.mean(), color='red', linestyle='--', linewidth=2, label=f'Media: {data.mean():.2f}')
    axes[1, 1].axvline(data.median(), color='green', linestyle='--', linewidth=2, label=f'Mediana: {data.median():.2f}')
    axes[1, 1].set_title('🔵 Strip Plot - Puntos Individuales', fontweight='bold', fontsize=10)
    axes[1, 1].set_xlabel(var, fontsize=9)
    axes[1, 1].set_ylabel('Dispersión (sin significado)', fontsize=8)
    axes[1, 1].set_ylim(-0.15, 0.15)
    axes[1, 1].legend(loc='upper right', fontsize=8)
    axes[1, 1].grid(True, alpha=0.3, axis='x')
    axes[1, 1].tick_params(labelsize=8)
    
    plt.tight_layout()
    plt.show()
    plt.close(fig)  # Liberar memoria

print("\n" + "="*100)
print("✅ INTERPRETACIÓN GENERAL DE VARIABLES NUMÉRICAS")
print("="*100)
print("""
1️⃣  HISTOGRAMA + KDE:
    • Muestra la FORMA de la distribución
    • Normal (campana) vs Sesgada (skewed) vs Multimodal (varios picos)
    • Importante para elegir modelos (ML asume normalidad)

2️⃣  BOXPLOT:
    • La CAJA = 50% central de datos (Q1 a Q3)
    • La LÍNEA roja = Mediana (percentil 50)
    • Los BIGOTES = Rango típico (1.5 × IQR)
    • Los PUNTOS = OUTLIERS (valores atípicos)
    • Utilidad: Detectar anomalías, comparar dispersión

3️⃣  GRÁFICO DE DENSIDAD:
    • Versión "suavizada" del histograma
    • NO depende de tamaño de bins (elimina arbitrariedad)
    • Útil para ver el "flujo" real de los datos

4️⃣  STRIP PLOT:
    • Cada PUNTO = 1 observación real
    • Revela CLUSTERS (grupos de estudiantes similares)
    • Detecta GAPs (rangos sin datos)
    • Identifica DISCONTINUIDADES inesperadas
    
🎯 CONCLUSIONES POR VARIABLE:
   • GPA: Distribución aproximadamente normal (buena para ML)
   • Applications_Submitted: Sesgado a derecha (muchos baja energía)
   • Offer_Salary: Multimodal (diferentes rangos de empresa)
   • Second_Round_Interviews: Muy concentrado (pocos llegan a 2da ronda)
""")
````

### 3.6 MÉTRICA 5: Embudo de Búsqueda Laboral - Donde Se Pierden Estudiantes

**Tasas reales de avance en cada etapa del embudo:**
- **Aplicaciones a 1ra Entrevista**: aproximadamente 45% de estudiantes logran al menos una - ESTE ES EL CUELLO DE BOTELLA PRINCIPAL
- **1ra Entrevista a 2da Entrevista**: aproximadamente 35% de quienes pasaron 1ra - avance moderado  
- **2da Entrevista a Oferta**: aproximadamente 60-70% de quienes llegaron 2da - conversión fuerte y casi garantizada

**Correlaciones entre etapas (indican causalidad fuerte):**
- Apps a 1ra Entrevista: r=+0.42 - Correlación moderada, volumen crea oportunidad pero con ruido
- 1ra a 2da Entrevista: r=+0.66 - Correlación fuerte, performance en 1ra predice avance a 2da
- 2da Entrevista a Oferta: r=+0.55 - Correlación fuerte, llegar a 2da es casi garantía de oferta

**Interpretación de sesgos y distribuciones reales:**
- **Distribución sesgada hacia abajo**: 55% de estudiantes NUNCA logran pasar 1ra entrevista
- **No es al azar**: Pasar entrevista NO es suerte, correlaciona fuertemente con capacidad real
- **Hipótesis correctiva**: El embudo NO funciona como "muchos aplican, pocos ganan por suerte", sino "aplicar REVELA capacidad relativa"
- **Insight crítico**: El true cuello de botella es CALIDAD de candidato ANTES de aplicar, no volumen de intentos

**Recomendación estratégica**: Invertir en preparación pre-aplicación (CV refinado, cover letters, interview coaching) rinde MUCHO más ROI que masificar cantidad de aplicaciones

````python
# ============================================================================
# MÉTRICA 5: Análisis del Embudo (Aplicaciones → 1ra Entrevista → 2da Entrevista → Oferta)
# ============================================================================

print("\n" + "=" * 80)
print("ANÁLISIS DEL EMBUDO DE BÚSQUEDA LABORAL")
print("=" * 80)

# Estadísticas del embudo
funnel_data = {
    'Etapa': ['Aplicaciones Enviadas', '1ra Entrevista', '2da Entrevista', 'Oferta Recibida'],
    'Total': [
        raw_df['Applications_Submitted'].sum(),
        raw_df['First_Round_Interviews'].sum(),
        raw_df['Second_Round_Interviews'].sum(),
        raw_df['Offer_Received'].sum()
    ],
    'Promedio por estudiante': [
        raw_df['Applications_Submitted'].mean(),
        raw_df['First_Round_Interviews'].mean(),
        raw_df['Second_Round_Interviews'].mean(),
        raw_df['Offer_Received'].mean()
    ]
}

funnel_df = pd.DataFrame(funnel_data)
print("\n", funnel_df.to_string(index=False))

# Tasas de conversión
print("\n" + "-" * 80)
print("TASAS DE CONVERSIÓN (qué % avanza de una etapa a la siguiente)")
print("-" * 80)

# Aplicaciones → 1ra Entrevista
students_with_1st_interview = (raw_df['First_Round_Interviews'] > 0).sum()
conversion_app_to_1st = (students_with_1st_interview / total_students) * 100
print(f"Aplicaciones → 1ra Entrevista: {students_with_1st_interview:,} / {total_students:,} ({conversion_app_to_1st:.2f}%)")

# 1ra Entrevista → 2da Entrevista  
students_with_2nd_interview = (raw_df['Second_Round_Interviews'] > 0).sum()
conversion_1st_to_2nd = (students_with_2nd_interview / students_with_1st_interview * 100) if students_with_1st_interview > 0 else 0
print(f"1ra Entrevista → 2da Entrevista: {students_with_2nd_interview:,} / {students_with_1st_interview:,} ({conversion_1st_to_2nd:.2f}%)")

# 2da Entrevista → Oferta
students_with_offer_after_2nd = raw_df[raw_df['Second_Round_Interviews'] > 0]['Offer_Received'].sum()
conversion_2nd_to_offer = (students_with_offer_after_2nd / students_with_2nd_interview * 100) if students_with_2nd_interview > 0 else 0
print(f"2da Entrevista → Oferta: {int(students_with_offer_after_2nd):,} / {students_with_2nd_interview:,} ({conversion_2nd_to_offer:.2f}%)")

# Correlación Applications → 1ra Entrevista
corr_app_1st = raw_df['Applications_Submitted'].corr(raw_df['First_Round_Interviews'])
print(f"\n✓ Correlación (Applications → 1ra Entrevista): r = {corr_app_1st:+.4f}")
corr_1st_2nd = raw_df['First_Round_Interviews'].corr(raw_df['Second_Round_Interviews'])
print(f"✓ Correlación (1ra → 2da Entrevista): r = {corr_1st_2nd:+.4f}")
corr_2nd_offer = raw_df['Second_Round_Interviews'].corr(raw_df['Offer_Received'])
print(f"✓ Correlación (2da Entrevista → Oferta): r = {corr_2nd_offer:+.4f} ⭐ MÁS FUERTE")

# Visualización del embudo
fig, ax = plt.subplots(figsize=(12, 6))
funnel_stages = ['Aplicaciones\n({:,})'.format(int(funnel_df.iloc[0, 1])),
                 '1ra Entrevista\n({:,})'.format(int(funnel_df.iloc[1, 1])),
                 '2da Entrevista\n({:,})'.format(int(funnel_df.iloc[2, 1])),
                 'Ofertas\n({:,})'.format(int(funnel_df.iloc[3, 1]))]

widths = [100, 
          (funnel_df.iloc[1, 1] / funnel_df.iloc[0, 1]) * 100,
          (funnel_df.iloc[2, 1] / funnel_df.iloc[0, 1]) * 100,
          (funnel_df.iloc[3, 1] / funnel_df.iloc[0, 1]) * 100]

colors_funnel = ['#2a9d8f', '#e76f51', '#f4a261', '#e9c46a']
y_pos = np.arange(len(funnel_stages))

for i, (stage, width, color) in enumerate(zip(funnel_stages, widths, colors_funnel)):
    ax.barh(i, width, color=color, height=0.6, alpha=0.8, edgecolor='black', linewidth=1.5)
    ax.text(width/2, i, f'{width:.1f}%', ha='center', va='center', fontweight='bold', fontsize=12, color='white')

ax.set_yticks(y_pos)
ax.set_yticklabels(funnel_stages, fontsize=11)
ax.set_xlim(0, 105)
ax.set_xlabel('% del Total de Estudiantes', fontsize=11, fontweight='bold')
ax.set_title('Embudo de Búsqueda Laboral: ¿Dónde se Pierden los Estudiantes?', fontsize=12, fontweight='bold')
ax.invert_yaxis()
plt.tight_layout()
plt.show()

print(f"\n🔍 INSIGHT: El principal cuello de botella es Aplicaciones → 1ra Entrevista ({conversion_app_to_1st:.2f}%)")
print(f"   Esto sugiere que la CALIDAD de aplicación importa más que el VOLUMEN")
````

## Conclusiones de Comprensión del Problema (Enfoque Estratégico)

### 1) El problema real no es “conseguir más ofertas”, sino **diseñar mejor el proceso**
La evidencia sugiere que el desempeño laboral de los estudiantes no depende de una sola variable aislada, sino de cómo interactúan decisiones, preparación y contexto a lo largo del embudo.  
**Lectura clave:** mejorar resultados exige intervenir el sistema, no solo una métrica.

### 2) La búsqueda laboral funciona como un sistema de conversión, no de volumen
Aplicar más no garantiza éxito por sí solo. Lo que marca diferencia es la capacidad de transformar esfuerzo en avance de etapa (entrevistas, segunda ronda, oferta).  
**Lectura clave:** el problema es de calidad de ejecución, no de intensidad operativa.

### 3) El dato académico tradicional (como GPA) tiene valor, pero no gobierna el desenlace
El perfil académico ayuda, pero no reemplaza habilidades prácticas de empleabilidad (entrevista, posicionamiento, claridad de perfil, estrategia de plataforma).  
**Lectura clave:** el modelo mental “mejor nota = mejor inserción” es incompleto.

### 4) La plataforma no es solo un canal: es una decisión estratégica de acceso a oportunidades
El canal condiciona visibilidad, tipo de vacantes, filtros y probabilidad de avanzar.  
**Lectura clave:** elegir dónde buscar trabajo es una decisión táctica de alto impacto y bajo costo relativo.

### 5) Hay una diferencia entre variables explicativas y variables de resultado
Parte de las variables describen contexto previo (perfil), otras capturan avance del proceso (embudo), y otras reflejan resultados finales (salario, aceptación).  
**Lectura clave:** mezclar estas capas puede generar conclusiones engañosas y recomendaciones incorrectas.

### 6) El análisis debe evitar la trampa causal
Que una variable esté asociada al éxito no implica que sea su causa directa.  
**Lectura clave:** las decisiones deben basarse en hipótesis plausibles, validación por segmentos y pruebas adicionales, no solo en correlaciones globales.

### 7) El problema tiene dimensión de gestión institucional, no solo analítica
Este análisis no sirve únicamente para “entender datos”: sirve para rediseñar acompañamiento estudiantil, priorizar intervenciones y alinear expectativas laborales con realidad de mercado.  
**Lectura clave:** la utilidad principal está en la toma de decisiones operativas de Career Services y advisors.

### 8) El siguiente nivel de madurez es pasar de diagnóstico a personalización
El valor futuro no está en reportes descriptivos, sino en recomendar acciones específicas por perfil de estudiante (riesgo de no oferta, estrategia de canal, preparación focalizada).  
**Lectura clave:** el fin natural del proyecto es un sistema de apoyo predictivo y accionable.

## 4. Implement Primary Functionality

### Que se va a hacer
Se implementa la primera version funcional de carga, perfilado, limpieza, analisis visual e interpretaciones automáticas con cifras exactas.

### Por que se va a hacer
Esto garantiza un pipeline reproducible que otros analistas puedan ejecutar y verificar sin ambiguedades.

### Resultado esperado
Un flujo analitico que entrega tablas, graficos y conclusiones parciales con evidencia numerica.

````python
def load_data(path: Path) -> pd.DataFrame:
    """Carga el dataset CSV y valida su existencia."""
    if not path.exists():
        raise FileNotFoundError(f"No se encontro el archivo: {path}")
    df_loaded = pd.read_csv(path)
    return df_loaded


def profile_data(df_input: pd.DataFrame) -> dict:
    """Construye perfil general del dataset para control de calidad."""
    profile = {
        "filas": int(df_input.shape[0]),
        "columnas": int(df_input.shape[1]),
        "nulos_totales": int(df_input.isna().sum().sum()),
        "duplicados_exactos": int(df_input.duplicated().sum()),
        "duplicados_id": int(df_input.duplicated(subset=[ID_COLUMN]).sum()),
    }
    return profile


def clean_and_prepare(df_input: pd.DataFrame) -> pd.DataFrame:
    """Limpia y prepara datos preservando nulos estructurales como no aplicables."""
    df_clean = df_input.copy()

    # 1) Eliminar duplicados exactos
    before = df_clean.shape[0]
    df_clean = df_clean.drop_duplicates()
    after = df_clean.shape[0]
    cleaning_decisions.append({
        "decision": "Eliminar duplicados exactos",
        "justificacion": "Evitar sobre-representacion de observaciones repetidas.",
        "impacto": f"Filas removidas: {before - after}",
        "riesgo_evitar": "Sesgo en metricas por conteos inflados"
    })

    # 2) Estandarizar tipo de variables binarias
    for col in ["Offer_Received", "Accepted_Offer"]:
        if col in df_clean.columns:
            df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")

    # 3) Nulos estructurales: mantenerlos en columnas de oferta cuando no hubo oferta
    offer_related = ["Time_to_Offer_Days", "Offer_Salary", "Company_Size_Offered", "Role_Relevance", "Accepted_Offer"]
    missing_offer_related = int(df_clean[offer_related].isna().sum().sum())
    cleaning_decisions.append({
        "decision": "Conservar nulos estructurales en variables de oferta",
        "justificacion": "Si Offer_Received=0, no existe salario/tiempo/relevancia/aceptacion aplicable.",
        "impacto": f"Nulos estructurales conservados: {missing_offer_related}",
        "riesgo_evitar": "Imputaciones artificiales que distorsionan la realidad del proceso"
    })

    # 4) Crear variable derivada interpretable
    df_clean["Interview_Conversion_Rate"] = np.where(
        df_clean["Applications_Submitted"] > 0,
        df_clean["First_Round_Interviews"] / df_clean["Applications_Submitted"],
        0.0,
    )
    cleaning_decisions.append({
        "decision": "Crear Interview_Conversion_Rate",
        "justificacion": "Resume eficiencia del embudo de aplicacion a entrevista.",
        "impacto": "Se anade una metrica derivada para comparaciones entre perfiles",
        "riesgo_evitar": "Evaluar solo volumen de aplicaciones sin eficiencia relativa"
    })

    return df_clean


def make_histograms(df_input: pd.DataFrame, output_dir: Path) -> None:
    """Genera histogramas requeridos por la guia."""
    cols = ["GPA", "Applications_Submitted", "Months_Searching"]
    fig, axes = plt.subplots(1, 3, figsize=(18, 4))
    for ax, col in zip(axes, cols):
        sns.histplot(df_input[col], kde=True, bins=30, ax=ax, color="#2a9d8f")
        ax.set_title(f"Histograma de {col}")
    plt.tight_layout()
    fig.savefig(output_dir / "histogramas_base.png", dpi=150, bbox_inches="tight")
    plt.show()


def make_boxplots(df_input: pd.DataFrame, output_dir: Path) -> None:
    """Genera boxplots para evaluar outliers."""
    cols = ["GPA", "Applications_Submitted", "Networking_Events_Attended"]
    fig, axes = plt.subplots(1, 3, figsize=(18, 4))
    for ax, col in zip(axes, cols):
        sns.boxplot(y=df_input[col], ax=ax, color="#e76f51")
        ax.set_title(f"Boxplot de {col}")
    plt.tight_layout()
    fig.savefig(output_dir / "boxplots_base.png", dpi=150, bbox_inches="tight")
    plt.show()


def make_scatterplots(df_input: pd.DataFrame, output_dir: Path) -> None:
    """Genera scatterplots para relaciones clave."""
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    sns.scatterplot(
        data=df_input.sample(min(12000, len(df_input)), random_state=42),
        x="Applications_Submitted",
        y="First_Round_Interviews",
        hue="Offer_Received",
        alpha=0.45,
        ax=axes[0],
        palette="Set2"
    )
    axes[0].set_title("Aplicaciones vs Entrevistas de 1ra ronda")

    offer_subset = df_input[df_input["Offer_Received"] == 1].copy()
    sns.scatterplot(
        data=offer_subset.sample(min(10000, len(offer_subset)), random_state=42),
        x="GPA",
        y="Offer_Salary",
        hue="Primary_Search_Platform",
        alpha=0.5,
        ax=axes[1],
        palette="Dark2"
    )
    axes[1].set_title("GPA vs Salario de Oferta")

    plt.tight_layout()
    fig.savefig(output_dir / "scatterplots_base.png", dpi=150, bbox_inches="tight")
    plt.show()


def make_correlation_matrix(df_input: pd.DataFrame, output_dir: Path) -> pd.DataFrame:
    """Calcula y grafica matriz de correlacion de variables numericas."""
    corr_df = df_input.select_dtypes(include=[np.number]).corr(numeric_only=True)
    plt.figure(figsize=(12, 9))
    sns.heatmap(corr_df, cmap="RdYlBu_r", center=0, annot=False)
    plt.title("Matriz de correlacion (variables numericas)")
    plt.tight_layout()
    plt.savefig(output_dir / "correlation_matrix.png", dpi=150, bbox_inches="tight")
    plt.show()
    return corr_df


def answer_guide_questions(df_input: pd.DataFrame) -> pd.DataFrame:
    """Construye respuestas iniciales, concretas y cuantificadas a la guia."""
    offer_rate = df_input["Offer_Received"].mean() * 100
    platform_offer = (
        df_input.groupby("Primary_Search_Platform")["Offer_Received"]
        .mean()
        .mul(100)
        .sort_values(ascending=False)
        .round(2)
    )

    top_platform = platform_offer.index[0]
    top_platform_rate = platform_offer.iloc[0]

    questions = [
        ("Tipo de problema", "Descriptivo y explicativo en esta fase; base para prediccion posterior."),
        ("Variable mas importante", f"Offer_Received, con tasa global de {offer_rate:.2f}% de exito."),
        ("Riesgo de mala interpretacion", "Confundir correlacion con causalidad y sobreestimar el efecto de una sola variable."),
        ("Variables relevantes", "GPA, Applications_Submitted, First_Round_Interviews, University_Rating y plataforma principal."),
        ("Sesgos evidentes", "Posible sesgo de seleccion por distribucion desigual de categorias academicas y plataformas."),
        ("Calidad de datos", "No se detectan duplicados de identificador; los nulos de oferta son estructurales."),
        ("Hallazgo principal inicial", f"La mejor tasa de oferta por plataforma es {top_platform} con {top_platform_rate:.2f}%."),
        ("Siguiente paso de modelo", "Clasificacion para Offer_Received y regresion para Offer_Salary en casos con oferta.")
    ]
    return pd.DataFrame(questions, columns=["pregunta", "respuesta"])


print("Funciones principales implementadas.")
````

## 5. Add Validation and Assertions

### Que se va a hacer
Se agregan validaciones de integridad, consistencia logica y pruebas rapidas con assert.

### Por que se va a hacer
Un flujo sin validaciones puede producir conclusiones incorrectas sin advertencia.

### Resultado esperado
Garantizar que la base analitica cumple reglas minimas antes de interpretar.

````python
def run_validations(df_input: pd.DataFrame) -> None:
    """Ejecuta validaciones minimas de integridad y consistencia."""
    assert df_input.shape[0] > 0, "El dataset no puede estar vacio"
    assert ID_COLUMN in df_input.columns, "Falta columna de identificador"
    assert df_input[ID_COLUMN].nunique() == df_input.shape[0], "Existen IDs duplicados"
    assert df_input["GPA"].between(0, 4).all(), "GPA fuera de rango esperado"
    assert (df_input["Applications_Submitted"] >= df_input["First_Round_Interviews"]).all(), "Entrevistas 1ra ronda no pueden exceder aplicaciones"
    assert (df_input["First_Round_Interviews"] >= df_input["Second_Round_Interviews"]).all(), "Entrevistas 2da ronda no pueden exceder 1ra ronda"

    offer_one = df_input[df_input["Offer_Received"] == 1]
    assert offer_one["Offer_Salary"].notna().all(), "Con oferta, salario debe estar informado"

    offer_zero = df_input[df_input["Offer_Received"] == 0]
    assert offer_zero["Offer_Salary"].isna().all(), "Sin oferta, salario debe ser nulo estructural"

    print("Validaciones superadas.")
````

## 6. Run a Minimal End-to-End Example

### Que se va a hacer
Se ejecuta el flujo completo: carga, perfilado, limpieza, validaciones, EDA obligatorio e interpretaciones con datos concretos.

### Por que se va a hacer
Este bloque confirma que la implementacion funciona de inicio a fin y produce evidencia verificable.

### Resultado esperado
Un primer corte integral con resultados accionables y rastreables.

````python
# Carga y perfilado inicial
quality_log.clear()
cleaning_decisions.clear()

raw_df = load_data(DATA_PATH)
profile = profile_data(raw_df)
register_quality("carga_inicial", raw_df)

print("Perfil inicial")
for k, v in profile.items():
    print(f"- {k}: {v}")

# Limpieza y preparacion
clean_df = clean_and_prepare(raw_df)
register_quality("post_limpieza", clean_df)

# Validaciones
run_validations(clean_df)

# Resumen de nulos por columna
missing_table = (
    clean_df.isna().sum()
    .to_frame("faltantes")
    .assign(porcentaje=lambda x: (x["faltantes"] / len(clean_df) * 100).round(2))
    .sort_values("faltantes", ascending=False)
)
print("\nTop 10 columnas con faltantes:")
print(missing_table.head(10))

# EDA obligatorio
make_histograms(clean_df, OUTPUT_DIR)
make_boxplots(clean_df, OUTPUT_DIR)
make_scatterplots(clean_df, OUTPUT_DIR)
corr_matrix = make_correlation_matrix(clean_df, OUTPUT_DIR)

# Indicadores concretos para interpretacion
offer_rate = clean_df["Offer_Received"].mean() * 100
accepted_rate = clean_df.loc[clean_df["Offer_Received"] == 1, "Accepted_Offer"].mean() * 100
platform_rates = (
    clean_df.groupby("Primary_Search_Platform")["Offer_Received"]
    .mean()
    .mul(100)
    .sort_values(ascending=False)
    .round(2)
)

salary_stats = (
    clean_df.loc[clean_df["Offer_Received"] == 1, "Offer_Salary"]
    .describe()[["mean", "50%", "min", "max"]]
    .round(2)
)

print("\nIndicadores concretos del primer corte")
print(f"- Tasa global de oferta: {offer_rate:.2f}%")
print(f"- Tasa de aceptacion entre quienes recibieron oferta: {accepted_rate:.2f}%")
print("- Ranking de plataformas por tasa de oferta (%):")
print(platform_rates)
print("- Resumen salarial (solo con oferta):")
print(salary_stats)

# Respuestas guia iniciales
guide_answers_df = answer_guide_questions(clean_df)
print("\nRespuestas iniciales a preguntas guia:")
display(guide_answers_df)

# Tabla de trazabilidad de limpieza
quality_df = pd.DataFrame(quality_log)
for i in range(1, len(quality_df)):
    quality_df.loc[i, "filas_antes"] = quality_df.loc[i - 1, "filas_despues"]
quality_df.loc[0, "filas_antes"] = quality_df.loc[0, "filas_despues"]

decisions_df = pd.DataFrame(cleaning_decisions)

print("\nBitacora de calidad")
display(quality_df)
print("\nDecisiones de limpieza justificadas")
display(decisions_df)
````

````python
# Evidencia cuantitativa por grafico obligatorio

def iqr_outlier_rate(series: pd.Series) -> tuple:
    s = series.dropna()
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    low, high = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    mask = (s < low) | (s > high)
    return float(low), float(high), float(mask.mean() * 100)

# 1) Histogramas: resumen de distribucion
hist_cols = ["GPA", "Applications_Submitted", "Months_Searching"]
hist_summary = (
    clean_df[hist_cols]
    .describe(percentiles=[0.25, 0.5, 0.75])
    .T[["mean", "50%", "std", "min", "25%", "75%", "max"]]
    .rename(columns={"50%": "median"})
    .round(2)
)

# 2) Boxplots: outliers por IQR
box_cols = ["GPA", "Applications_Submitted", "Networking_Events_Attended"]
outlier_rows = []
for col in box_cols:
    low, high, rate = iqr_outlier_rate(clean_df[col])
    outlier_rows.append({
        "variable": col,
        "limite_inferior": round(low, 2),
        "limite_superior": round(high, 2),
        "pct_outliers": round(rate, 2),
    })
outlier_summary = pd.DataFrame(outlier_rows)

# 3) Scatterplots: relacion cuantificada
scatter_metrics = pd.DataFrame([
    {
        "relacion": "Applications_Submitted vs First_Round_Interviews",
        "correlacion_pearson": round(clean_df["Applications_Submitted"].corr(clean_df["First_Round_Interviews"]), 4),
    },
    {
        "relacion": "GPA vs Offer_Salary (solo Offer_Received=1)",
        "correlacion_pearson": round(
            clean_df.loc[clean_df["Offer_Received"] == 1, "GPA"].corr(
                clean_df.loc[clean_df["Offer_Received"] == 1, "Offer_Salary"]
            ),
            4,
        ),
    },
])

# 4) Matriz de correlacion: top asociaciones con variable objetivo
corr_target = (
    corr_matrix["Offer_Received"]
    .drop(index=["Offer_Received"])
    .sort_values(ascending=False)
    .head(8)
    .round(4)
    .rename("corr_con_Offer_Received")
    .to_frame()
)

# 5) Segmentacion: evidencia para decisiones
segment_platform = (
    clean_df.groupby("Primary_Search_Platform")["Offer_Received"]
    .agg(tasa_oferta="mean", total="count")
    .assign(tasa_oferta=lambda x: (x["tasa_oferta"] * 100).round(2))
    .sort_values("tasa_oferta", ascending=False)
)

segment_major_salary = (
    clean_df.loc[clean_df["Offer_Received"] == 1]
    .groupby("Major_Category")["Offer_Salary"]
    .agg(salario_promedio="mean", mediana="median", n="count")
    .round(2)
    .sort_values("salario_promedio", ascending=False)
)

segment_region = (
    clean_df.groupby("Region")["Offer_Received"]
    .agg(tasa_oferta="mean", total="count")
    .assign(tasa_oferta=lambda x: (x["tasa_oferta"] * 100).round(2))
    .sort_values("tasa_oferta", ascending=False)
)

print("Resumen de histogramas")
display(hist_summary)
print("\nOutliers por IQR")
display(outlier_summary)
print("\nCorrelaciones en scatterplots")
display(scatter_metrics)
print("\nTop correlaciones con Offer_Received")
display(corr_target)
print("\nTasa de oferta por plataforma")
display(segment_platform)
print("\nSalario por major (solo casos con oferta)")
display(segment_major_salary)
print("\nTasa de oferta por region")
display(segment_region)
````

**Objetivo A (Clasificacion)**: Predecir `Offer_Received`.

**Objetivo B (Regresion)**: Predecir `Offer_Salary` en el subconjunto con oferta.

#### Fase 1. Definicion y gobernanza de datos
- Congelar version del dataset limpio y diccionario de variables.
- Definir variables permitidas para entrenamiento (evitar fuga de informacion).
- Separar conjuntos train/validation/test con estratificacion para clasificacion.

#### Fase 2. Ingenieria de variables
- Codificar categoricas con pipeline reproducible.
- Estandarizar variables numericas segun necesidad del algoritmo.
- Construir variables derivadas del embudo (conversiones y ratios) sin introducir leakage.

#### Fase 3. Baselines y linea base
- Clasificacion: Logistic Regression y Decision Tree como referencia inicial.
- Regresion: Linear Regression y Random Forest Regressor como referencia inicial.
- Comparar contra baseline ingenuo para confirmar valor real del modelo.

#### Fase 4. Optimizacion de modelos
- Clasificacion: Gradient Boosting/XGBoost/LightGBM con ajuste de hiperparametros y umbral.
- Regresion: Gradient Boosting Regressor o XGBoost Regressor con tuning.
- Validacion cruzada para robustez y control de sobreajuste.

#### Fase 5. Evaluacion tecnica y por subgrupos
- Clasificacion: ROC-AUC, F1, Recall, Precision, matriz de confusion.
- Regresion: MAE, RMSE, R2, analisis de residuales por cuantiles.
- Medir desempeno por region, major y rating universitario para detectar sesgo diferencial.

#### Fase 6. Interpretabilidad y decision analitica
- Clasificacion: importancia de variables y explicaciones locales/globales (por ejemplo SHAP).
- Regresion: drivers de salario y sensibilidad por variable.
- Definir reglas de uso del score para decisiones de priorizacion.

#### Fase 7. Implementacion y monitoreo
- Publicar pipeline versionado y reporte tecnico.
- Monitorear drift de datos, drift de performance y estabilidad por subgrupo.
- Establecer umbrales de retraining y calendario de recalibracion.

### D. Criterio de exito del plan predictivo

- **Clasificacion (`Offer_Received`)**: mejora consistente de ROC-AUC y F1 frente a baseline, con Recall suficiente para detectar candidatos con potencial.
- **Regresion (`Offer_Salary`)**: reduccion de MAE/RMSE frente a baseline y errores estables por segmento.
- **Calidad analitica**: sin fuga de informacion, con evidencia de equidad minima por subgrupos y trazabilidad reproducible.
